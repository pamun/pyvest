# Investment in Fama-French portfolios

```python
import pandas as pd
import matplotlib.pyplot as plt

from pyvest import Investment
```

## Define start month and end month

```python
start_date="2021-01"
end_date="2024-08"
```

## Create an Investment object

* The investment amount is 1000 dollars.
* The amount is assumed to be invested on the last trading day of the month.
* The weights correspond to the initial weights. The portfolios is not automatically rebalanced.
* Dividends are automatically reinvested.

```python
investment = Investment(["Mkt", "beta_4", "value_1", "momentum_10", "size_3"], 
                        1000,
                        weights={"Mkt": 0.15, "beta_4": 0.25, "value_1": 0.2, 
                                 "momentum_10": 0.05, "size_3": 0.35}, 
                        start_date=start_date, 
                        end_date=end_date)
```

## Plots

```python
investment.plot_profit()
```

```python
investment.plot_return()
```

```python
investment.plot_value()
```

```python
investment.plot_weight()
```

# Create an investment that is monthly rebalanced

```python
investment_rebalanced = Investment(["Mkt", "beta_4", "value_1", "momentum_10", "size_3"], 
                                   1000, 
                                   weights={"Mkt": 0.15, "beta_4": 0.25, "value_1": 0.2, 
                                            "momentum_10": 0.05, "size_3": 0.35}, 
                                   start_date=start_date, 
                                   end_date=end_date, 
                                   rebalance=True)
```

# Plot the two investments on the same graph

```python
investment.plot_profit(investment_rebalanced)
```

```python
investment.plot_return(investment_rebalanced)
```

```python
investment.plot_value(investment_rebalanced)
```

```python
investment.plot_weight(investment_rebalanced)
```
