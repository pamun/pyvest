```python
from pyvest import Investment
```

## Define start date and end date

```python
start_date = "2021-01-01"
end_date = "2024-06-30"
```

## Create an Investment object

```python
investment = Investment(["MCD", "MSFT"], 
                        6000, 
                        start_date=start_date, 
                        end_date=end_date)
```

## Make transactions

### Buy transactions

```python
investment.buy("MCD", quantity=10, date="2021-01-04")
```

```python
investment.buy("MSFT", quantity=15, date="2022-01-05")
```

### Sell transactions

```python
investment.sell("MCD", quantity=5, date="2022-04-04")
```

```python
investment.sell("MSFT", quantity=10, date="2023-01-06")
```

## Plots

```python
investment.plot_profit()
```

```python
investment.plot_return()
```

```python
investment.plot_value()
```

```python
investment.plot_weight()
```

## Plots with transactions

```python
investment.plot_profit(show_transactions=True)
```

```python
investment.plot_return(show_transactions=True)
```

```python
investment.plot_value(show_transactions=True)
```

```python
investment.plot_weight(show_transactions=True)
```
