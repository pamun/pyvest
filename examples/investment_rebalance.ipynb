# Investment in stocks based on weights

```python
cd ../../
```

```python
from pyvest import Investment
```

## Define start date and end date

```python
start_date = "2021-01-01"
end_date = "2024-08-01"
```

## Create an Investment object

* The investment amount is 1000 dollars.
* The stocks are bought according to the weights on the start date.
* The weights correspond to the initial weights. The portfolios is not automatically rebalanced.

```python
investment = Investment(["MCD", "MSFT"], 
                        1000, 
                        weights={"MCD": 0.6, "MSFT": 0.4},
                        start_date=start_date, 
                        end_date=end_date)
```

## Plots

```python
investment.plot_profit()
```

```python
investment.plot_return()
```

```python
investment.plot_value()
```

```python
investment.plot_weight()
```

# Create an investment that is monthly rebalanced

* The investment is rebalanced on the last trading day of each month according to the weights.

```python
investment_rebalanced = Investment(["MCD", "MSFT"], 
                                   1000,  
                                   weights={"MCD": 0.6, "MSFT": 0.4}, 
                                   start_date=start_date, 
                                   end_date=end_date, 
                                   rebalance=True)
```

# Plot the two investments on the same graph

```python
investment.plot_profit(investment_rebalanced)
```

```python
investment.plot_return(investment_rebalanced)
```

```python
investment.plot_value(investment_rebalanced)
```

```python
investment.plot_weight(investment_rebalanced)
```
